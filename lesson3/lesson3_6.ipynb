```python
nums = [1, 2, 3, 4, 5, 6]
#透過索引取值
print(nums[0])
print(nums[5])

#透過slice取值
print(nums[0:3])
print(nums[:3])
print(nums[-3:-1])
print(nums[-3:])

for num in nums:
    print(num)
```
